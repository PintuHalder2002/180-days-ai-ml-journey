## 1. List Comprehension

```python
numbers = [1 , 2 , 3 , 4 , 5]

sqr_numbers = []

for number in numbers:
    sqr_numbers.append(number**2)

print(sqr_numbers)
```

```python
sqr_numbers2 = [n*n for n in numbers]

sqr_numbers2
```

```python
import time

long_list = range(10000000)
sqr_list = []

start = time.time()
for n in long_list:
    sqr_list.append(n*n)
end = time.time()
print("Time taken by for loop:", (end-start))
```

```python
start = time.time()
sqr_list = [n*n for n in long_list]
end = time.time()
print("Time taken by list comprehension:", (end-start))
```

```python
price = [100 , 120 , 80 , 50]
earning_per_share = [9 , 15 , 40 , 2]
```

```python
tuple(zip(price , earning_per_share))
```

```python
pe = []

start = time.time()

for p , e in zip(price , earning_per_share):
    pe.append(round(p/e , 2))

end = time.time()

print(pe)

print("Time taken" , end - start)
```

```python
start = time.time()

pe = [round(p/e , 2) for p , e in zip(price , earning_per_share)]

end = time.time()

print(pe)

print("Time taken" , end - start)
```

## 2. If else condition in list comprehension

```python
price = [100 , 120 , 80 , 50]
earning_per_share = [0 , 3 , 6 ,8]

pe = [p/e if e != 0 else -1 for p, e in zip(price , earning_per_share)]

```

```python
words = ["apple" , "banana" , "guava" , "orange"]

lengths = [len(word) for word in words]

print(lengths)
```

```python
prices = [120 , 99 , 150 , 80 , 200]

new_prices = [price*0.9 if price > 100 else price for price in prices]

print(new_prices)
```

## 3. Dictionary comprehension

```python
# using first principle of dictionary

names = ["Adam" , "Marry" , "Pintu" , "Mammam"]
ages = [32 , 29 , 23 , 23]

new_dict  = dict()

for i , j in zip(names , ages):
    new_dict[i] = j

print(new_dict)
```

```python
# using dictionary comprehension

name_age_dict = {names[i]:ages[i] for i in range(len(names))}

print(name_age_dict)
```

## 4. Set Comprehension

```python
transactions = [
    {'customer_name': "Mira" , "amount": 56 , "date": "2024-08-01"},
    {'customer_name': "Mira" , "amount": 12.3 , "date": "2024-09-05"},
    {'customer_name': "Mohan" , "amount": 21 , "date": "2024-09-03"},
    {'customer_name': "Mohan" , "amount": 36.8 , "date": "2024-09-10"},
    {'customer_name': "Lucy" , "amount": 66.3 , "date": "2024-09-10"}
]
```

```python
# using set comprehension

unique_customers_set = {trans['customer_name'] for trans in transactions}

print(unique_customers_set)
```

## Q3. How do you use a dictionary comprehension to invert keys and values of a given dictionary

```python
{v:k for k , v in {'a':1 , 'b':2}.items()}
```
