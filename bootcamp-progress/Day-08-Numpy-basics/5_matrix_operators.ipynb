```python
import numpy as np
# Sales data for Quarter 1 (Matrix1)
# Rows represent different products in quantities, columns represent different regions.

q1 = np.array([
    [200, 220, 250], # Product A
    [150, 180, 210], # Product B
    [300, 330, 360]  # Product C
])


q2 = np.array([
    [209, 231, 259], # Product A
    [155, 192, 222], # Product B
    [310, 340, 375]  # Product C
])
```

```python
q1 + q2
```

```python
q2 - q1
```

```python
((q2-q1)/q1)*100
```

```python
prices = np.array([
    [10, 12, 11], # prices for Product A in different regions
    [8, 9, 10],   # prices for Product B in different regions
    [15, 16, 17]  # prices for Product C in different regions
])

q1_revenue = q1 * prices
q1_revenue # total revenue for q1 in different countries
```

```python
# Discount
q1_discounts = q1_revenue * 0.2
q1_discounts
```

```python
net_revenue = q1_revenue - q1_discounts
net_revenue
```

```python
print(np.sum(net_revenue))
```
