```python
# JSON string
json_string = '''
{
    "name": "Maulik",
    "age": 30,
    "is_employee": true,
    "skills": ["Python", "Data Analysis", "Machine Learning"],
    "address": {
        "street": "123 Main st",
        "city": "Wonderland"
    }
}
'''


#ij json boolean values must be in lower case.

print(type(json_string))
```

```python
import json

record = json.loads(json_string)
print(record)
```

```python
print(type(record))
```

```python
print(record['age'])
```

```python
data = {
    "name": "Bob",
    "age": 25,
    "is_employee": False,
    "skills": ["Javascript", "Web Development"],
    "address": {
        "street": "456 Elm st",
        "city": "Springfield"
    }
}
```

```python
print(type(data))
```

```python
data_str = json.dumps(data)

print(data_str,"\n")

print(type(data_str))
```

```python
with open("output.json" , "w") as f:
    json.dump(data, f, indent = 4)
```

```python
with open("products.json" , "r") as f:
    products = json.load(f)


products
```

```python
type(products)
```

```python
products_list = [{'product_name': product['name'], 'product_price': product['price']} for product in products]

products_list
```

```python
# Malformed JSON STRING

json_string = '''
{
    "name": "Krishna",
    "age": 30         
    "is_employee": True
}
'''

# "," is missing.
```

```python
json.loads(json_string)
```

```python
try:
    json.loads(json_string)
except json.JSONDecodeError as e:
    print(f"Failed to parse JSON: {e}")
```
